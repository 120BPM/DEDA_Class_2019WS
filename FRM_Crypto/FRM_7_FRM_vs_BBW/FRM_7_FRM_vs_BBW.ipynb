```python
import numpy as np
import pandas as pd
import plotly
import plotly.graph_objs as go
import matplotlib.pyplot as plt
```

## Compare weighted BBW and FRM ws. 48

```python
lambdas = pd.read_csv('lambdas.csv', index_col = 'time')
lambdas.index = pd.to_datetime(lambdas.index)

BBW = pd.read_csv('BBW.csv')
BBW['time'] = pd.to_datetime(BBW['time'])
```

```python
fig = plt.figure(figsize=[12,6])
ax = fig.add_subplot(111)
plt.plot(lambdas.index,lambdas[lambdas<2], label = 'NReturn_ws48',color='k', linewidth =0.5,alpha=0.8)
plt.plot(pd.to_datetime(BBW.time),BBW['weighted BBW'],label = 'BBW_weighted',color='r',alpha=1)
plt.grid(axis='y')

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)

plt.legend(prop={'size': 8},frameon=False,loc='upper right')
plt.savefig('FRM with weighted BBW.png',transparent=True)
```

## boxplot of individual Lambdas

```python
ind_lambdas = pd.read_csv('ind_lambdas.csv')
ind_lambdas=ind_lambdas.iloc[:,1:]
```

```python
ind_lambdas.shape
```

```python
ind_lambdas.head()
```

```python
window=7*24 # hours for a week
```

```python
#25 weeks
round(len(ind_lambdas)/window)-1
```

```python
ind_lambdas.loc[window*0:window*(0+1)-1,:]
```

```python
# make a boxplot based on individual lambda value in a week, regard every individual lambdas from different coins indifferent 
mx=pd.DataFrame()
for week in range(round(len(ind_lambdas)/window)-1):
    mx[week]=pd.Series(ind_lambdas.loc[window*week:window*(week+1)-1,:].values.flatten())
```

```python
# nr. of columns is the nr. of weeks, each column contains 2016(=12coins*24hours*7days) lambda values for a single week from 12 coins

print(mx.shape)
mx.head()
```

```python
fig=plt.figure(figsize=[45,15])
ax=fig.add_subplot(111)
_=plt.boxplot(mx.T)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

ax.grid(axis = 'y')
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)

plt.savefig('./boxplot.png',dpi=400, transparent = True)
```

```python
# cut extrem values for a better view
fig=plt.figure(figsize=[45,15])
ax=fig.add_subplot(111)
_=plt.boxplot(mx.T)

plt.ylim(0,0.6)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

ax.grid(axis = 'y')
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)

plt.savefig('./boxplot_cut.png',dpi=400, transparent = True)
```
